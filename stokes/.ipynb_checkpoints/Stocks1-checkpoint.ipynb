```python
# Define Starting Params
stocksymbols=['AAPL','GOOG','TSLA','AMZN']
start_date='20110101'
end_date='20170801'
```

```python
#Get Data from the Internet
def Get_Dict(stocksymbols,start_date,end_date):
    
    tickername=','.join(stocksymbols)
    #URL for get from Api
    url=''.join(['https://www.quandl.com/api/v3/datatables/WIKI/PRICES.json?date.gte=',
             start_date,
             '&date.lt=',
             end_date,
             '&ticker=',
             tickername,
             '&api_key=Sfmbqrscecuo5Qaxbr5Y'])
    requesteddata=requests.get(url)
    mydict=requesteddata.json()
    return mydict
```

```python
def Get_DataFrame(stocksymbols,start_date,end_date):
    
    mydict=Get_Dict(stocksymbols,start_date,end_date)
    
    tickers=[]
    dates=[]
    opens=[]
    highs=[]
    lows=[]
    close=[]
    volume=[]

    for each in mydict['datatable']['data']:
        tickers.append(each[0])
        dates.append(each[1])
        opens.append(each[9])
        highs.append(each[10])
        lows.append(each[11])
        close.append(each[12])
        volume.append(each[13])
    
    mydataframe=pd.DataFrame({
    'ticker':tickers,
    'dates':dates,
    'opens':opens,
    'highs':highs,
    'lows':lows,
    'close':close,
    'volume':volume})
    
    new=pd.to_datetime(mydataframe.dates)
    timetuples=[datetime.date.timetuple(x) for x in new]
    datesnew=[datetime.date(x.tm_year,x.tm_mon,x.tm_mday) for x in timetuples]
    mydataframe['dates']=datesnew
    mydataframe.set_index('dates',inplace=True)
    
    return mydataframe
```

```python
orig_df=Get_DataFrame(stocksymbols,start_date,end_date)
```

```python
def GetReturn(dataframe,days):
    
    dataframe.sort_index(inplace=True)
    _pctchange=dataframe.close.pct_change(periods=days)*100
    dataframe['returnovernext'+str(days)]=_pctchange.shift(-1*days)
    return dataframe

def GetMovingAvg(dataframe,period):
    dataframe['movavg'+str(period)]=dataframe.close.rolling(window=period,min_periods=period).mean()
    return dataframe
```

```python
def Build_Features(mydataframe):
# How was the close relative to the highs and the lows of the day. Smaller the value the closer it closed to the high or the low
#Another observation is if the range is small the values are insignificant and maybe we need to give them lesser significance
#     mydataframe['close_rel_low']=(mydataframe['close'] - mydataframe['lows'])*100/mydataframe['close']
#     mydataframe['close_rel_high']=(mydataframe['highs'] - mydataframe['close'] )*100 /mydataframe['close']
    
    mydataframe['range']=mydataframe['highs']-mydataframe['lows']
    mydataframe=mydataframe.groupby('ticker').apply(GetMovingAvg,50)
    mydataframe=mydataframe.groupby('ticker').apply(GetMovingAvg,100)
    mydataframe=mydataframe.groupby('ticker').apply(GetMovingAvg,200)
    mydataframe=mydataframe.groupby('ticker').apply(GetReturn,1)
    mydataframe=mydataframe.groupby('ticker').apply(GetReturn,5)
    mydataframe=mydataframe.groupby('ticker').apply(GetReturn,30)
    
    mydataframe['distfrom50MA']=(mydataframe['close']-mydataframe['movavg50'])*100/mydataframe['close']
    mydataframe['distfrom100MA']=(mydataframe['close']-mydataframe['movavg100'])*100/mydataframe['close']
    mydataframe['distfrom200MA']=(mydataframe['close']-mydataframe['movavg200'])*100/mydataframe['close']
#     mydataframe['zeroline']=0
#     mydataframe['minus10line']=-10
#     mydataframe['minus15line']=-15
#     mydataframe['minus20line']=-20

    return mydataframe
```
