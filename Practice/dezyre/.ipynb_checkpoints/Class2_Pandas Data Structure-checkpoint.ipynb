```python
colstats=[(column,df[column].isnull().sum(),df[column].notnull().sum()) for column in list(df.columns)]
```

```python
removecols=[]
for col in colstats:
    if np.round(100*col[1]/(col[1]+col[2]),decimals=2)>25:
        print 'Dropping ',col[0],'\t\t',np.round(100*col[1]/(col[1]+col[2]),decimals=2)
        removecols.append(col[0])
```

```python
df.groupby(['Complaint Type']).count().sort(columns=['Unique Key'],axis=0,ascending=False)
# df.sort(columns=['Location Type'],axis=0)
# help(df.sort)
```

```python
removecols
df.drop([x for x in removecols],axis=1,inplace=True)
```

```python
df.head()
```

```python
# df1=df.isnull().any()
# emptycols=df1[df1.values==True].index.tolist()
imputecols=df.isnull().any().index[df.isnull().any().values==True].tolist()
colsandtypeslist=dict(zip(df.dtypes.index.tolist(),df.dtypes.values.tolist()))
imputedict={}
for col in imputecols:
    if col in colsandtypeslist.keys():
        imputedict[col]=str(colsandtypeslist[col])
imputedict
```

```python
df['Address Type'].mode()[0]
```

```python
df['Address Type']=df['Address Type'].fillna(df['Address Type'].mode()[0])
```

```python
df['Address Type'].value_counts()
```

```python
df['Address Type'].isnull().sum()
```

```python
for col in imputedict.keys():
    if imputedict[col]=='object':
        df[col].fillna(df[col].mode()[0],inplace=True)
    elif imputedict[col]=='float64':
        df[col].fillna(df[col].mean(),inplace=True)
```
