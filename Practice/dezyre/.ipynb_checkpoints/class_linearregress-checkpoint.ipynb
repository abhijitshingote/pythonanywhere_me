```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
#Create x variable
x=np.random.uniform(0,5,1000)
plt.plot(xrange(1000),x)
plt.show()
```

```python
#Create Y variable
#mu and sigma for the noise
mu=2
sigma=0.5
y=4*x + np.random.normal(mu,sigma,1000)
plt.plot(xrange(1000),y)
plt.show()
```

```python
print x.mean(),y.mean()
plt.scatter(x,y)
plt.show()
```

```python
#Second example
from numpy import arange,array,ones
from pylab import plot,show
from scipy import stats
import scipy,scipy.stats
from scipy import stats
```

```python
dataset='http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv'
```

```python
data=pd.read_csv(dataset,index_col=0)
data.head()
```

```python
#what is the ROI of $1 spend on TV,Radio,Newspaper
```

```python
for col in data.columns[:-1]:
    x=stats.linregress(data[col],data.Sales)
    print col,'-',x.slope*1+x.intercept
```

```python
fig,axs=plt.subplots(1,3,sharey=True)
data.plot(kind='scatter',x='TV',y='Sales',ax=axs[0],figsize=(17,5))
data.plot(kind='scatter',x='Radio',y='Sales',ax=axs[1])
data.plot(kind='scatter',x='Newspaper',y='Sales',ax=axs[2])
```
