```python
from sklearn import datasets, linear_model
from sklearn import cross_validation as cv
from scipy import linalg
```

```python
# predictive model creation
xtrain = modeldata_train.iloc[:,0:(modeldata.shape[1]-1)].as_matrix()
ytrain = modeldata_train['target']

xtest = modeldata_test.iloc[:,0:(modeldata.shape[1]-1)].as_matrix()
ytest = modeldata_test['target']
```

```python
xtrain.shape, xtest.shape
```

```python
model1 = linear_model.LinearRegression(normalize=True)
model1.fit(xtrain,ytrain)
```

```python
print ('coefficients: \n', model1.coef_)
```

```python
def adjustedrsquare(model, x, y):
    yhat = model.predict(x)
    n, k = x.shape
    sse = np.sum(np.square(yhat-y),axis=0)
    sst = np.sum(np.square(y-np.mean(y)),axis=0)
    R2 = 1 - sse/sst
    adjR2 = R2 - (1-R2) * (float(k)/(n-k-1))
    return R2, adjR2
```

```python
model1.adjR2, model1.R2 = adjustedrsquare(model1,xtrain, ytrain)
print model1.adjR2, model1.R2
```

```python
# Null: the beta coefficients are = zero
# Alt: the beta coefficients are != zero
```

```python
yhat = model1.predict(xtest)
plt.scatter(ytest,yhat)
plt.plot(ytest,ytest)
plt.show()
```

```python
def ReturnPValue(model, X, Y):
    Yhat = model.predict(X)
    n, k = X.shape
    sse = np.sum(np.square(Yhat-Y),axis=0)
    x = np.hstack((np.ones((n,1)),np.matrix(X)))
    df = float(n-k-1)
    sampleVar = sse/df
    sampleVarianceX = x.T*x
    covarianceMatrix = linalg.sqrtm(sampleVar*sampleVarianceX.T)
    se = covarianceMatrix.diagonal()[1:]
    betasTStat = np.zeros(len(se))
    for i in xrange(len(se)):
        betasTStat[i]=model.coef_[i]/se[i]
        betasPValue = 1-stats.t.cdf(abs(betasTStat),df)
        return betasPValue
```

```python
# computing the pvalues from the model using a custom function
# covariance = (Sum ( x-xbar)(y-ybar))/(n-1)

ReturnPValue(model1,xtrain,ytrain)
```

```python
def adjustedrsquare(model, x, y):
    yhat = model.predict(x)
    n, k = x.shape
    sse = np.sum(np.square(yhat-y),axis=0)
    sst = np.sum(np.square(y-np.mean(y)),axis=0)
    R2 = 1 - sse/sst
    adjR2 = R2 - (1-R2) * (float(k)/(n-k-1))
    return R2, adjR2
```

```python
def ErrorMeasures(model, X, Y):
    Yhat = model.predict(X)
    MAPE = np.mean(abs(Y-Yhat)/Y)*100
    MSSE = np.mean(np.square(Y-Yhat))
    return MAPE, MSSE
```

```python
regr.MAPE, regr.MSSE = ErrorMeasures(regr, XTest,YTest)
print regr.MAPE, regr.MSSE
```

```python
from sklearn import datasets, linear_model
from sklearn import cross_validation as cv
from scipy import linalg
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
boston = datasets.load_boston()
scaler = StandardScaler()
```

```python
boston['data'].shape
```

```python
# transform the dataset using standard scaler
X = scaler.fit_transform(boston['data'])

poly = PolynomialFeatures(2)
X = poly.fit_transform(X)
```

```python
print X
```

```python
Y = boston['target']
```

```python
# Y = alpha + beta1*X1 + beta2*X1^2 + beta3*X2 + beta4*X2^2..........
```

```python
featurenames = ['intercept'] + list(boston['feature_names'])

polyFeaturenames = [];

for i, x in enumerate(featurenames):
    for y in featurenames[i:]:
        if(x == 'intercept'):
            polyFeaturenames.append(y)
        elif(x==y):
            polyFeaturenames.append((y+'_Square'))
        else:
            polyFeaturenames.append((x+'_'+y))
    
```

```python
print len(polyFeaturenames)
```

```python
X.shape
```

```python
regr = linear_model.LinearRegression(normalize=True)
regr.fit(X,Y)
```

```python
regr.MAPE, regr.MSSE = ErrorMeasures(regr, X,Y)
print regr.MAPE, regr.MSSE
```

```python
#18.6410673886 26.7081911773 # with categorical predictors
# 18.2411582259 27.3707637199 # linear regression without categorical predictors
#9.37133332736 5.98141972603 # including polynomial features
```

```python
regr.R2, regr.AdjR2 = adjustedrsquare(regr,X,Y)
print regr.AdjR2, regr.R2
```
