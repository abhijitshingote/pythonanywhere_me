# Outliers

```python
#Create a dataframe with dates as your index
States=['NY','NY','NY','NY','FL','FL','GA','GA','FL','FL']
data=[1.0,2,3,4,5,6,7,8,9,10]
idx=pd.date_range('1/1/2012',periods=10,freq='MS')
df1=pd.DataFrame(data, index=idx, columns=['Revenue'])
df1['State']=States

#Create a second dataframe
data2=[10.0,10,9,9,8,8,7,7,6,6]
idx2=pd.date_range('1/1/2013', periods=10, freq='MS')
df2=pd.DataFrame(data2,index=idx2,columns=['Revenue'])
df2['State']=States

#Combine dataframes
df=pd.concat([df1,df2])
df
```

```python
#Assumption : We consider that 95% of our data is not of the outlier kind.(this is mostly for normally or gaussian distributed data)
```

```python
df.Revenue.hist()
```

```python
#Identify outliers using first method ie. mean plus minus 1.96 covers 95% of normally dist values
newdf=df.copy()
newdf.Revenue[1:3]=-99 #test
newdf['mean - rev']=newdf.Revenue - newdf.Revenue.mean()
stddev=1.96*newdf['Revenue'].std()
newdf['Outlier']=(abs(newdf['mean - rev'])) > stddev
newdf
```

```python
#Method 2 using lambda. this time we will check outliers at the state level
# Not sure what is happening here - how does x - x.mean() know that operation is to be performed
# on the Revenue field without explicitly stating x.Revenue

newdf=df.copy()
StateMonth= newdf.groupby(['State',lambda x:x.month])
newdf['Outlier']=StateMonth.transform( lambda x: abs(x-x.mean()) > 1.96*x.std())
newdf
```

```python
#Method 3 - using apply function. Apply function only -
# Remember we dont have to specify that we re picking the 2nd object by saying group[1]
#Conversely if you are just iterating over a groupby object , you will have specify u want to look at group[1],
# which is the second element in the tuple.
# Instead , all of the values within the group are directly available to us.
#Question - When apply applies the outlier function, where are the group by indexes gone
#We can see those  indexes when we apply the np.mean fuinction(NY,FL,GA)

newdf=df.copy()
State=newdf.groupby(['State'])

def outlier(group):
    group['Outlier']=group.Revenue - group.Revenue.mean() > 1.96 * group.Revenue.std()
    return group

print State.apply(outlier)
print State.apply(np.mean)
```

```python
# Digression - This is to show that apply will return the number of rows that are 
# dependent on the function that we are applying
#If you return the original group - return group - apply function returns a dataframe
#If you return a expression, apply function returns a series
#Note- Even when the series is returned , no of ROWS in the series depends on the type of aplly function that is applied

print df.groupby(['State']).apply(np.mean)

# or 
def rev_sub_mean(group):
    return group.Revenue - group.Revenue.mean() >1.96*group.Revenue.std()
def rev_sub_mean2(group):
    return sum(group.Revenue.values)

print df.groupby(['State']).apply(rev_sub_mean)
print df.groupby(['State']).apply(rev_sub_mean2)
print type(df.groupby(['State']).apply(rev_sub_mean))
print type(df.groupby(['State']).apply(rev_sub_mean2))
```

```python
#Above Methods are valid if a dataset is normally distributed. If it isnt, then we should opt for
# quantile distribution
# To find Quantiles, arrange data in ascending order, number that falls in the 25% or 75% "POSITION" are the 
#  25 & 75 percent quantile
#Inter quartile Range = 75% Quantile - 25% Quantile
# Outlier is anything that is  more than (1.5 times * InterQuantile Range ) Far Away from the 25% & 75 % quantiles

```

```python
# Outliers with Quantile function using lambda - Again not sure how lambda uses the Revenue value by default
newdf1=df.copy()
State1 = newdf1.groupby(['State'])
newdf1['Lower_than'] = State1.transform( lambda x : x.quantile(q=.25) - 1.5*(x.quantile(q=.75) - x.quantile(q=.25)))
newdf1['Upper_than'] = State1.transform( lambda x : x.quantile(q=.75) + 1.5*(x.quantile(q=.75) - x.quantile(q=.25)))
newdf1
newdf1['Outlier']= (newdf1['Revenue']< newdf1['Lower_than']) | (newdf1['Revenue'] > newdf1['Upper_than'])
newdf1
```

```python
#Outliers with Quantile using basic method
newdf=df.copy()
#Calculate Quantiles and InterQuantile Range
newdf['Lower_limit']=newdf.Revenue.quantile(q=.25) - 1.5*(newdf.Revenue.quantile(q=.75) - newdf.Revenue.quantile(q=.25))
newdf['Upper_limit']=newdf.Revenue.quantile(q=.75) + 1.5*(newdf.Revenue.quantile(q=.75) - newdf.Revenue.quantile(q=.25))
newdf['Outlier']=(newdf['Revenue']<newdf['Lower_limit']) | (newdf['Revenue']>newdf['Upper_limit'])
newdf
```

```python
#Outliers with Quantile using basic method but grouped at State
newdf=df.copy()
State=newdf.groupby(['State'])

newdf5=pd.DataFrame()

def Outlier(group):
    group['Lower_limit']=group.Revenue.quantile(q=.25) - 1.5*(group.Revenue.quantile(q=.75) - group.Revenue.quantile(q=.25))
    group['Upper_limit']=group.Revenue.quantile(q=.75) + 1.5*(group.Revenue.quantile(q=.75) - group.Revenue.quantile(q=.25))
    group['outlier']= (group['Revenue']<group['Lower_limit']) | (group['Revenue']>group['Upper_limit'])
    return group

for group in State:
    newgroup=Outlier(group[1])
    newdf5=newdf5.append(newgroup)
    
newdf5
```

```python
#Outliers with Quantile using Apply method
newdf=df.copy()
State=newdf.groupby(['State'])
# State=newdf.groupby(['State',lambda x: x.month])   Alternatively if you wanted to group by at State,Month
def Outlier(group):
    group['Lower_limit']=group.Revenue.quantile(q=.25) - 1.5*(group.Revenue.quantile(q=.75) - group.Revenue.quantile(q=.25))
    group['Upper_limit']=group.Revenue.quantile(q=.75) + 1.5*(group.Revenue.quantile(q=.75) - group.Revenue.quantile(q=.25))
    group['outlier']= (group['Revenue']<group['Lower_limit']) | (group['Revenue']>group['Upper_limit'])
    return group
newdf2=State.apply(Outlier)
newdf2
```

# Merge and Join

```python
import pandas as pd

df1=pd.DataFrame({'Name':['Student 1','Student 2'],
                  'MathScore': [85,95]})
df2=pd.DataFrame({'Name': ['Student 3','Student 4'],
                  'MathScore': [80,90]})
df3=pd.DataFrame({'Name': ['Student 1','Student 2'],
                 'ScienceScore': [90,100]})
```

```python
newdf=df1.append(df2)   #df1 doesnt get affected. If you save it, assign it to new variable.Also index should be redone
newdf
```

```python
newdf.index=range(len(newdf))
newdf
```

```python
#If we wanted to add a partner name and their math score - this gets matched based on index
newdf_=df1.copy()
newdf_['Partner Name']=df2.Name
newdf_['Partner Score']=df2.MathScore
newdf_
```

```python
#Merge doesnt look at index. If "how" isnt given  or 'on isnt given , it searches for a field on its own
df1.merge(df3) #automatically picked Name
```

```python
df1.merge(df3,on='Name')
```
